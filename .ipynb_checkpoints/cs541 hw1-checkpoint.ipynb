```python
import random

def run_a1(num_dirt = 1):
    start = random.randint(0,8)

    room = [[0 for x in range(3)] for y in range(3)]
    dirt_loc = random.sample(range(0,9), num_dirt)
    dirt_xloc = [x//3 for x in dirt_loc]
    dirt_yloc = [x%3 for x in dirt_loc]

    score = 0
    for d in range(num_dirt):
        room[dirt_xloc[d]][dirt_yloc[d]] = 1

    print(room)
    cur = (start//3, start%3)
    print(cur)
    steps = 0
    acts = ['s','n','l','r','u','d']
    while score < num_dirt and steps < 10000:
        x_pos = cur[0]
        y_pos = cur[1]
        action = acts[random.randint(0,5)]
        if room[y_pos][x_pos] ==1 and action == 's':
                room[y_pos][x_pos] = 0
                score += 1
                if score < num_dirt:
                    continue
                else:
                    break
        elif action == 'l' and cur[0] > 0:
            cur = (cur[0]-1, cur[1])
        elif action == 'r' and cur[0] < 2:
            cur = (cur[0]+1, cur[1])
        elif action == 'u' and cur[1] > 0:
            cur = (cur[0], cur[1]-1)
        elif action == 'd' and cur[1] < 2:
            cur = (cur[0], cur[1]+1)
        else:
            continue

        steps += 1
    return steps

print(room)

print(score,steps)
a1_perf = []
for i in range(100):
    a1_perf.append(run_a1(1))
print(sum(a1_perf)/100)
```
